```python
unigrams_df = pd.read_csv('unigrams.txt', sep='\t', names=['Unigram', 'Count'])
totalUnigrams = len(unigrams_df)
bigrams_df = pd.read_csv('bigrams.txt', sep='\t', names=['Bigram', 'Count'])
totalBigrams = len(bigrams_df)

print("Number of unigrams: " + str(totalUnigrams) + " Number of Bigrams: " + str(totalBigrams))
print("\n")
display(unigrams_df.head(100), bigrams_df.head(100))
```

Complete the function below such that it computes the probability of a given sentence using the unigram and bigram counts from (a). Do not apply add-one smoothing

```python
def probability(sentence):

    words = sentence.split()
    sentenceProbability = 1
    totalWords = unigrams_df['Count'].sum()
    print("Sentence:" +str(words) + "\n")

    for i in range(len(words) - 1):
        unigram_count = unigrams_df[unigrams_df['Unigram'] == words[i]]['Count']
        if not unigram_count.empty:
            unigram_count = int(unigram_count)
        else:
            unigram_count = 0
        print("Unigram count for " + words[i] + " = " + str(unigram_count))

        bigram = words[i] + " " + words[i+1]
        bigram_count = bigrams_df[bigrams_df['Bigram'] == bigram]['Count']
        if not bigram_count.empty:
            bigram_count = int(bigram_count)
        else:
            bigram_count = 0

        #Bigram probability
        bigram_prob = bigram_count / unigram_count if unigram_count > 0 else 0
        sentenceProbability *= bigram_prob

        print("Bigram count for " + bigram + " = " + str(bigram_count))
        print("Bigram probability for " + bigram + " = " + str(bigram_prob))
        print("-----------------------------------------------------------------\n")

    print("Sentence Probability: " + str(sentenceProbability) + "\n")
    return sentenceProbability

prob_i_love_you = probability("i love you")
prob_i_hate_you = probability("i hate you")
print(prob_i_love_you > prob_i_hate_you)
```

```python
def probability_with_smoothing(sentence):

    words = sentence.split()
    log_sentence_probability = 0
    totalWords = unigrams_df['Count'].sum()
    print("Sentence:" + str(words) + "\n")
    uniqueWords = len(unigrams_df)


    for i in range(len(words) - 1):

        unigram_count = unigrams_df[unigrams_df['Unigram'] == words[i]]['Count']
        unigram_count = int(unigram_count) if not unigram_count.empty else 0

        smoothed_unigram_count = unigram_count+1
        print("Unigram count for " + words[i] + " = " + str(smoothed_unigram_count))

        bigram = words[i] + " " + words[i+1]

        bigram_count = bigrams_df[bigrams_df['Bigram'] == bigram]['Count']
        bigram_count = int(bigram_count) if not bigram_count.empty else 0
        smoothed_bigram_count = bigram_count+1

        #probability of the bigram with smoothing
        bigram_prob = smoothed_bigram_count / (smoothed_unigram_count + uniqueWords)

        #probability to log space to prevent underflow
        log_bigram_prob = math.log(bigram_prob)

        log_sentence_probability += log_bigram_prob

        print("Bigram count for " + bigram + " = " + str(smoothed_bigram_count))
        print("Bigram probability for " + bigram + " = " + str(bigram_prob))
        print("-----------------------------------------------------------------\n")

    print("Log Sentence Probability: " + str(log_sentence_probability) + "\n")
    return log_sentence_probability

log_i_love_you = probability_with_smoothing("i love you")
log_i_hate_you = probability_with_smoothing("i hate you")

print(log_i_love_you > log_i_hate_you)
```

#Task 3
Using the bigram counts (bigrams_df), write a function that implements the Shannon visualization method:


```python
def ShannonVisualisation(seed="<S>", max_length=100):

    bigram_probs = {}
    for _, row in bigrams_df.iterrows():
        bigram, count = row['Bigram'], row['Count']
        first_word, second_word = bigram.split()
        if first_word not in bigram_probs:
            bigram_probs[first_word] = {second_word: count}
        else:
            bigram_probs[first_word][second_word] = bigram_probs[first_word].get(second_word, 0) + count

    #Convert counts to probabilities
    for first_word in bigram_probs:
        total_count = sum(bigram_probs[first_word].values())
        for second_word in bigram_probs[first_word]:
            bigram_probs[first_word][second_word] /= total_count

    #Generate text
    current_word = seed
    generated_text = []
    while current_word != "</S>" and len(generated_text) < max_length:
        next_words_probs = bigram_probs.get(current_word, None)
        print("Current word: " + current_word + "\n")

        #break if no next words
        if not next_words_probs:
            break

        words, probabilities = zip(*next_words_probs.items())
        current_word = np.random.choice(words, p=probabilities)
        generated_text.append(current_word)

    return ' '.join(generated_text).strip()
```
