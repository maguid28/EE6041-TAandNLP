# Task 1 : Complete the following code snippet to download Peter Norvig’s unigram corpus, and load it into a dataframe, unigrams_df

Write a simple spelling correction function that takes a nonWord string as input (i.e. non-word misspelling such as “acress”), and prints out a list of candidate corrections.


```python
!wget "https://norvig.com/ngrams/count_1w.txt"
```

```python
unigrams_df = pd.read_csv('count_1w.txt', sep='\t', names=['Word', 'Count'])
```

```python
display(unigrams_df.head(100))
```

```python
def nonWordSpellingCorrection(nonWord):
    candidates = []

    total_frequency = unigrams_df['Count'].sum()

    for index, row in unigrams_df.iterrows():
        word = row['Word']

        if isinstance(word, str):
            distance = jellyfish.damerau_levenshtein_distance(nonWord, word)

            if distance <= 2:
                frequency = row['Count']
                probability = frequency / total_frequency
                candidates.append((word, frequency, probability))

    candidates_df = pd.DataFrame(candidates, columns=['Word', 'Count', 'P(word)'])

    candidates_df = candidates_df.sort_values(by='Count', ascending=False)

    return candidates_df
```

```python
word = input("Provide an incorrectly spelled word: ")
candidates_df = nonWordSpellingCorrection(word)
print(f"Misspelled Word = {word}")
display(candidates_df)
```

# Task 2 : Use the TextBlob Library (https://textblob.readthedocs.io/en/dev/) to write a function that takes a string (e.g.  acress) as input and returns a list of (word, confidence) tuples with spelling suggestions.


```python
def spelling_correction(input_string):
    blob = TextBlob(input_string)

    corrected_string = blob.correct()

    word = Word(input_string)
    candidates = word.spellcheck()

    return corrected_string, candidates

input_string = input("Provide an incorrect word: ")
corrected_string, candidates = spelling_correction(input_string)
print("Provided Word = " + input_string)
print("Candidates = " + str(candidates))
```

#Task 3 : Use the TextBlob Library (https://textblob.readthedocs.io/en/dev/) to write a function that takes a sentence as input and corrects its spelling.


```python
def correct_spelling(sentence):
    words = sentence.split()
    corrected_sentence = ""

    for word in words:
      corrected_word = str(TextBlob(word).correct())
      corrected_sentence+= corrected_word + " "

    return corrected_sentence

input_sentence = input("Enter a sentence to correct spelling: \n")

corrected_sentence = correct_spelling(input_sentence)
print("Original: " + str(input_sentence) + "\n")
print("Corrected: " + str(corrected_sentence) + "\n")
```
