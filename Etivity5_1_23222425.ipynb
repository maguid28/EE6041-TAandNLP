```python
import rarfile

def extract_rar(rar_path, output_path):
    with rarfile.RarFile(rar_path) as opened_rar:
        opened_rar.extractall(path=output_path)

rar_path = 'opinion-lexicon-English.rar'
output_path = '..'
extract_rar(rar_path, output_path)
```

```python
positive_words_file_path = '/opinion-lexicon-English/positive-words.txt'
negative_words_file_path = '/opinion-lexicon-English/negative-words.txt'

positive_words = {}
negative_words = {}

def load_words(file_path, words_dict, polarity):
    with open(file_path, 'r', encoding='ISO-8859-1') as file:
        for line in file:
            if line.startswith(';') or not line.strip():
                continue
            word = line.strip()

            if polarity == "pos": words_dict[word] = 1
            else : words_dict[word] = 0

load_words(positive_words_file_path, positive_words, "pos")
load_words(negative_words_file_path, negative_words, "neg")

print(positive_words)
```

```python
def calculate_prior_probabilities(dataset):
    total_samples = len(dataset)
    positive_samples = sum(label for _, label in dataset)
    negative_samples = total_samples - positive_samples

    prior_probability_positive = positive_samples / total_samples
    prior_probability_negative = negative_samples / total_samples

    return prior_probability_positive, prior_probability_negative


def calculate_likelihoods_with_laplace_smoothing(dataset):
    positive_samples = [text for text, label in dataset if label == 1]
    print(positive_samples)
    negative_samples = [text for text, label in dataset if label == 0]
    print(negative_samples)

    positive_word_counts = Counter(word for text in positive_samples for word in text.split())
    negative_word_counts = Counter(word for text in negative_samples for word in text.split())

    unique_words = set(positive_word_counts.keys()).union(set(negative_word_counts.keys()))
    vocab_size = len(unique_words)

    laplace_smoothing_factor = 1

    positive_total_words = sum(positive_word_counts.values())
    negative_total_words = sum(negative_word_counts.values())

    positive_likelihoods = {word: (positive_word_counts[word] + laplace_smoothing_factor) /
                            (positive_total_words + vocab_size * laplace_smoothing_factor)
                            for word in unique_words}
    negative_likelihoods = {word: (negative_word_counts[word] + laplace_smoothing_factor) /
                            (negative_total_words + vocab_size * laplace_smoothing_factor)
                            for word in unique_words}

    return positive_likelihoods, negative_likelihoods
```

```python
prior_probabilities = calculate_prior_probabilities(my_dataset)
likelihoods = calculate_likelihoods_with_laplace_smoothing(my_dataset)
```

```python
def calc_posterior_probability(new_sample, prior_probabilities, likelihoods):
    positive_likelihoods, negative_likelihoods = likelihoods
    prior_probability_positive, prior_probability_negative = prior_probabilities

    words = new_sample.split()
    positive_prob = np.log(prior_probability_positive)
    negative_prob = np.log(prior_probability_negative)

    for word in words:
        positive_prob += np.log(positive_likelihoods.get(word, 1 / (sum(positive_likelihoods.values()) + len(positive_likelihoods))))
        negative_prob += np.log(negative_likelihoods.get(word, 1 / (sum(negative_likelihoods.values()) + len(negative_likelihoods))))

    positive_prob = np.exp(positive_prob)
    negative_prob = np.exp(negative_prob)
    total_prob = positive_prob + negative_prob
    posterior_probability_positive = positive_prob / total_prob
    posterior_probability_negative = negative_prob / total_prob

    return posterior_probability_positive, posterior_probability_negative
```

```python
def sentimentAnalyzer(sentence, prior_probabilities, likelihoods) :
  posterior_probability_positive, posterior_probability_negative = calc_posterior_probability(sentence, prior_probabilities, likelihoods)

  print(f"String = {sentence}")

  if posterior_probability_positive > posterior_probability_negative :
    sentiment = "Positive Sentiment"
  else :
    sentiment = "Negative Sentiment"

  print(sentiment)
  print(f'Positive Confidence: {posterior_probability_positive}')
  print(f'Negative Confidence: {posterior_probability_negative}')
  print("------------------------------------------------------")


sentimentAnalyzer('NLP is bad', prior_probabilities, likelihoods)
sentimentAnalyzer('Happy', prior_probabilities, likelihoods)
sentimentAnalyzer('Ugly and Happy', prior_probabilities, likelihoods)
sentimentAnalyzer('Cat', prior_probabilities, likelihoods)
```

```python
def sentiment_analysis_rule_based(text):
    # Count positive and negative word appearances in the text
    positiveCount = sum(1 for word in text.lower().split() if word.lower().strip() in positive_words)
    negativeCount = sum(1 for word in text.lower().split() if word.lower().strip() in negative_words)

    total_count = positiveCount + negativeCount

    print(f"String = {text} \n")

    # Determining sentiment based on counts
    if positiveCount > negativeCount:
        print(f"Polarity: Positive")
        print(f"Confidence: {positiveCount/total_count}")
    elif positiveCount < negativeCount:
        print(f"Polarity: Negative")
        print(f"Confidence: {negativeCount/total_count}")
    else:
        print(f"Polarity: Neutral")
        print(f"Confidence: {positiveCount/total_count}")
```
