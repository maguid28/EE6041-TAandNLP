# Task 4

Write a function that takes sets of training and testing documents as input and uses the textblob.classifiers module to:
 a) train a Naive Bayes classifier using the training set
 b) test it using the test set and print out the accuracy of the model.


## Training Set

```python
trainingSet = [
    ("London is the Capital of GB", 'GB'),
    ("Oxford is a city in GB", 'GB'),
    ("Dublin is the capital of Ireland", 'IE'),
    ("Limerick is a city in Ireland", 'IE'),
]
```

## Test Set

```python
testSet = [
    ("University of Limerick", "IE"),
    ("University College Dublin", "IE"),
    ("Imperial College London", "GB"),
    ("University of Oxford", "GB"),
    ("Ireland & GB", "GB"),
]
```

## Train

```python
cl = NaiveBayesClassifier(trainingSet)
```

## Test

```python
accuracy = 0
correct = 0
all = 0
for doc in testSet:
  print("--------------------------------------------------------------------------------")
  print(f"\nTest Document: {doc[0]} \n")

  prob_dist = cl.prob_classify(doc[0])
  prob_dist.max()
  print(f'Predicted Class: {prob_dist.max()}')

  if prob_dist.max() == doc[1] :
      correct +=1

  all+=1

accuracy = correct / all
print("--------------------------------------------------------------------------------")
print(f'\nAccuracy: {accuracy}')
```

## Imports

```python
import requests
import pandas as pd
from textblob import TextBlob
import nltk
nltk.download('punkt')
from textblob.classifiers import NaiveBayesClassifier
```

```python
!wget "https://storage.googleapis.com/dataset-uploader/bbc/bbc-text.csv"

data = pd.read_csv('bbc-text.csv')

data.head()
```

## Documents Per Category

```python
techcount = 0
businesscount = 0
sportcount = 0
politicscount = 0
entertainmentcount = 0

for row in data['category'] :
  if row == 'tech':
    techcount+=1
  elif row == 'business':
    businesscount+=1
  elif row == 'sport':
    sportcount+=1
  elif row == 'politics':
    politicscount+=1
  elif row == 'entertainment':
    entertainmentcount+=1
```

## Split Data

```python
data = data.sample(frac=1).reset_index(drop=True)
split_index = int(len(data) * 0.8)
train_data = data[:split_index]
test_data = data[split_index:]
```

## Train Set Samples

```python
train_data.head()
```

## Test Set Samples

```python
test_data.head()
```

## DataFrame to list of tuples


```python
test_tuples = []
train_tuples = []

for index, row in train_data.iterrows():
    text = row['text']
    category = row['category']
    train_tuples.append((text, category))

for index, row in test_data.iterrows():
    text = row['text']
    category = row['category']
    test_tuples.append((text, category))
```

## Train Classifier

```python
cl = NaiveBayesClassifier(train_tuples)
```

## Test Classifier

```python
correct = 0
total = len(test_tuples)
for text, label in test_tuples:
    predicted = cl.classify(text)
    if predicted == label:
        correct += 1

accuracy = correct / total
```

## Results

```python
print(f"Number of Documents: {data.shape[0]}\n")

print('Documents per category: \n')

print(f'Tech: {techcount}')
print(f'Business: {businesscount}')
print(f'Sport: {sportcount}')
print(f'Politics: {politicscount}')
print(f'Entertainment: {entertainmentcount}')

print(f'\nTrain Size: {len(train_tuples)}')
print(f'Test Size: {len(test_tuples)}\n')

print(f'Train Set Samples:\n')
print(f'{train_data.head()}\n')
print(f'Test Set Samples:\n')
print(f'{test_data.head()}\n')

print("\nAccuracy:", accuracy)
```
